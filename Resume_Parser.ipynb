```python
import fitz
def extract_text_pdf(path):
    text=""
    pdf= fitz.open(path)
    for page in pdf:
        text=text+page.get_text()
    return text
resume_text=extract_text_pdf(r'MelissaSequeira_resume_July25.pdf')
print(resume_text)
```

```python
import re

def lower_text(text):
    text=text.lower()
    return text
low_resume_text=lower_text(resume_text)
print(low_resume_text)
```

```python
email=re.findall(r'[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+[A-Za-z]{2,}',low_resume_text)
phone=re.findall(r'\+?\d[\d -]{8,}\d',low_resume_text)
print(email)
print(phone)
```

```python
clean_text=re.sub(r'[^\w\s]','',low_resume_text)
clean_text=re.sub(r'\s+',' ',clean_text)
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

en_stopwords=stopwords.words('english')

def remove_stopwords(text):
    words=' '.join([word for word in clean_text.split()])
    return words

rem_stopw_resume_text=remove_stopwords(clean_text)
print(rem_stopw_resume_text)
```

```python
from nltk.tokenize import word_tokenize
nltk.download('punkt')
def token_word(text):
    tokened=word_tokenize(text)
    return tokened
tokened_text=token_word(rem_stopw_resume_text)
print(tokened_text)
```

```python
from nltk.stem import WordNetLemmatizer
lem=WordNetLemmatizer()
def lemmatize_text(text):
    words=text.split()
    lemme_text=[]
    for word in words:
        word=lem.lemmatize(word, pos='v')
        lemme_text.append(word)
    return lemme_text
lemmatize_res_text=lemmatize_text(resume_text)  
print(lemmatize_res_text)
```

```python
import spacy

nlp = spacy.load('en_core_web_sm')

# Keep original text for better NER results
doc = nlp(low_resume_text.title())

name = None
for ent in doc.ents:
    if ent.label_ == "PERSON":
        name = ent.text
        print("Detected Name:", name)
        break

        
```

```python
skills_list=["Python", "Java", "Machine Learning", "SQL", "Django", "Keras", "NLP"]
found_skills=[]
for skill in skills_list:
    if skill in resume_text:
        found_skills.append(skill)
print(found_skills)
```

```python
parsed_resume={
    "name":name,
    "email":email,
    "phone":phone,
    "skills":[skill for skill in found_skills] 


}
for key, value in parsed_resume.items():
    print(f"{key}: {value}")
```
