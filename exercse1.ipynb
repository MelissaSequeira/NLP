| Pattern  | Matches                                   | Example                                     |        |                               |
| -------- | ----------------------------------------- | ------------------------------------------- | ------ | ----------------------------- |
| `.`      | Any single character except newline       | `a.c` → matches `"abc"`, `"axc"`            |        |                               |
| `\d`     | Digit (0–9)                               | `\d\d` → matches `"12"`                     |        |                               |
| `\D`     | Non-digit                                 | `\D+` → `"abc"` from `"abc123"`             |        |                               |
| `\w`     | Word character (`a–z`, `A–Z`, `0–9`, `_`) | `\w+` → `"Hello_123"`                       |        |                               |
| `\W`     | Non-word character                        | `\W+` → `"!"`, `" "`                        |        |                               |
| `\s`     | Whitespace (space, tab, newline)          | `\s+` → `"   "`                             |        |                               |
| `\S`     | Non-whitespace                            | `\S+` → `"Hello"`                           |        |                               |
| `^`      | Start of string                           | `^Hello` matches `"Hello World"`            |        |                               |
| `$`      | End of string                             | `world$` matches `"Hello world"`            |        |                               |
| `[...]`  | Character set                             | `[aeiou]` matches vowels                    |        |                               |
| `[^...]` | Negated character set                     | `[^0-9]` matches non-digits                 |        |                               |
| \`       | \`                                        | OR                                          | \`(cat | 

| `*`      | 0 or more repetitions                     | `a*` matches `""`, `"a"`, `"aaa"`           |        |                               |
| `+`      | 1 or more repetitions                     | `a+` matches `"a"`, `"aa"`                  |        |                               |
| `?`      | 0 or 1 repetition                         | `colou?r` matches `"color"` and `"colour"`  |        |                               |
| `{n}`    | Exactly n repetitions                     | `\d{4}` matches `"2024"`                    |        |                               |
| `{n,}`   | n or more repetitions                     | `\d{2,}` matches `"12"`, `"1234"`           |        |                               |
| `{n,m}`  | Between n and m repetitions               | `\d{2,4}` matches `"12"`, `"123"`, `"1234"` |        |                               |

TOKENIZATION

```python
import nltk
nltk.download('punkt')
nltk.download("punkt_tab")
from nltk.tokenize import sent_tokenize , word_tokenize
sentence="I am finding for someone loyal. I expect same thing from me"
print(sent_tokenize(sentence))
print(word_tokenize(sentence))
```

STEMING

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()

words=["Melissa", "walking", "studied" ,"beautiful","flower", "foodie"]

for word in words:
    print(word + "-->"+ ps.stem(word) + "\n")
```

LEMMATIZATION

```python
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')

lem=WordNetLemmatizer()
words=["Melissa", "walking", "studied" ,"beautiful","flower", "foodie"]


for t in words:
    print(t+ "-->" +lem.lemmatize(t, pos='v')+"\n")
    
```

```python
nlp=spacy.load("en_core_web_sm")
emma_ja = "emma woodhouse handsome clever and rich with a comfortable home and happy disposition seemed to unite some ..."
spacy_doc=nlp(emma_ja)
```

```python
pos_doc=pd.DataFrame(columns=["token","pos_tag"])
```

```python
for token in spacy_doc:
    pos_doc=pd.concat([pos_doc, pd.DataFrame.from_records([{'token':token.text,'pos_tag':token.pos_}])], ignore_index=True)
    
```

```python
from spacy import displacy
from spacy import tokenizer
import re
```

```python
my_text="Hello, my name is Melissa. I am a student at the University of Michigan. I am interested in data science and machine learning."
print(my_text)
```

```python
spacy_doc=nlp(my_text)
```

```python
for word in spacy_doc.ents:
    print(word.text, word.label_)
```

```python
displacy.render(spacy_doc, style="ent", jupyter=True)
```

```python
clean_text=re.sub(r'[^\w\s]','',my_text).lower()
print(clean_text)
```

```python
spacy_doc_clean=nlp(clean_text)
```

```python
for word in spacy_doc_clean.ents:
    print(word.text, word.label_)
```

```python
displacy.render(spacy_doc_clean,style='ent',jupyter=True, options={'distance': 100})
```
