Bag of Words
A bag of words is a representation of text that describes the occurrence of words in a document. The bag of words representation is often used as a feature vector for machine learning algorithms.

```python
%pip install scikit-learn
from sklearn.feature_extraction.text import CountVectorizer
```

```python
corpus = [
    "The cat sat on the mat",
    "The dog barked at the cat"
]
```

```python
vet=CountVectorizer()
vetTrans=vet.fit_transform(corpus)
```

```python
vocab = vet.get_feature_names_out() #vet koi use karke vocabulary from corpus nikala
print(vocab)
```

```python
print("array of word count:",vetTrans.toarray()) #arraya of word count
```

```python
# TF-IDF (Term Frequency-Inverse Document Frequency) is a statistical measure used to evaluate how important a word is to a document in a collection or corpus.
# It increases proportionally to the number of times a word appears in the document but is offset by the frequency of the word in the corpus.
# This helps to adjust for the fact that some words appear more frequently in general.

# In summary:
# - Term Frequency (TF): Measures how frequently a term occurs in a document.
# - Inverse Document Frequency (IDF): Measures how important a term is by reducing the weight of terms that occur in many documents.
# - TF-IDF: The product of TF and IDF, giving a score that highlights words that are frequent in a document but not across all documents.
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

```

```python
corpus={
    "The cat sat on the mat",
    "The dog barked at the cat"
}
```

```python
tfidf=TfidfVectorizer()
tfdifTrans=tfidf.fit_transform(corpus)
```

```python
vocab=tfidf.get_feature_names_out() #tfidf koi use karke vocabulary from corpus nikala
print(vocab)
```

```python
print("freq and imp count of words:",tfdifTrans.toarray()) #freq and imp count of words
```
